```python
import pandas as pd
```

```python
df = pd.read_csv('../dados/titanic.csv')
df
```

```python
#Questão 1
#Quantas pessoas sobreviveram ao acidente?
print(str(df.loc[(df['sobreviveu'] == 'sim')].shape[0]) + ' pessoas sobreviveram ao acidente')
df[(df['sobreviveu'] == 'sim')]
#Sobreviveram 342 pessoas
```

```python
#Questão 2
#Quantas criancas menores de 12 anos sobreviveram ao acidente?
print(str(df.loc[(df['idade'] < 12.0 ) & (df['sobreviveu'] == 'sim')].shape[0]) + ' crianças menores de 12 anos sobreviveram')
df.loc[(df['idade'] < 12.0 ) & (df['sobreviveu'] == 'sim')]
#39 crianças menores de 12 anos sobreviveram
```

```python
#Questao 3
#Quantas pessoas embarcaram nos portos de Southampton OU Queenstow?
southampton = df.loc[(df['embarque'] == 'Southampton')].shape[0]
queenstow = df.loc[(df['embarque'] == 'Queenstow')].shape[0]
total = southampton + queenstow
print('{} pessoas embarcaram nos portos de Southampton ou Queenstow, sendo que: \n' .format(total))
print('{} pessoas embarcaram em Southampton' .format(southampton))
print('{} pessoas embarcaram em Queenstow' .format(queenstow))
#721 pessoas embarcaram nos portos de Southampton ou Queenstow, sendo que: 

#644 pessoas embarcaram em Southampton
#77 pessoas embarcaram em Queenstow
```

```python
#Questao 4
#Crie um objeto chamado passageiras que seleciona somente as passageiras.
passageiras = df.loc[(df['sexo'] == 'feminino')]
print(passageiras)
#Objeto "passageiras" criado e contém a seleção das 314 passageiras da tabela.
```

```python
#Questão 5
#Crie um objeto chamado criancas_cherbourg que seleciona as crianças com menos de 12 anos E que embarcaram na cidade de Cherbourg.
criancas_cherbourg = df.loc[(df['idade'] < 12.0 ) & (df['embarque'] == 'Cherbourg')]
print(criancas_cherbourg)
#Objeto "criancas_cherbourg" criado e contém a seleção das 10 crianças menores de 12 anos que embarcaram em Cherbourg da tabela.
```

```python
#Questão 6
#Crie um objeto chamado titanic_menor e selecione as colunas de nome, idade, classe e embarque
nome = df['nome']
idade = df['idade']
classe = df['classe']
embarque = df['embarque']
titanic_menor = [[nome], [idade], [classe], [embarque]]
print(titanic_menor)
#Objeto "titanic_menor" criado com as colunas nome, idade, classe e embarque selecionadas.
```

```python
#Questão 7
#Supondo que o valor da tarifa no dataset está em libras, e que 1£ = R$ 6.43, crie uma nova coluna chamada tarifa_real que contenha a tarifa de embarque em reais
df2 = df.copy() # faco uma cópia do objeto
df2['tarifa_real'] = df2['tarifa'] * 6.43
df2.head()
#Coluna "tarifa_real" criada com os valores da tarifa de embarque em reais.
```

```python
#Questão 8
#Agrupe pela coluna sobreviveu e calcule o número de passageiros por grupo (sim/nao)
df[['id_passageiro', 'sobreviveu']].groupby(['sobreviveu']).count()
#Podemos visualizar que 549 não sobreviveram e 342 sobreviveram
```

```python
#Questão 9
#Agrupe pelo local de embarque e calcule a média e a mediana da tarifa de cada grupo.
#df[['embarque', 'tarifa']].groupby(['embarque', 'tarifa']).count()
df.agg({'embarque':['count'],
              'tarifa':['count', 'max', 'min', 'mean']})
#Tarifa máxima:     262375.000000
#Tarifa mínima:     0.000000
#Tarifa média:      4026.270117
```

```python
#Questão 10
#Agrupe pela classe e conte a quantidade de pessoas por classe
df3 = df.copy()
def classifica_passageiro(row_value):
    if row_value == 3:
        social = 'terceira classe'
    elif row_value == 2:
        social = 'segunda classe'
    else:
        social = 'primeira classe'
    return social

df3['tipo_social'] = df['classe'].apply(classifica_passageiro)
df3[['tipo_social', 'classe']].groupby(['tipo_social']).count()
#Agrupamento por classe realizado, trazendo a quantidade de pessoas por classe:
#primeira classe	216
#segunda classe     184
#terceira classe	491

```

```python
#Questão 11
#Agrupe pela classe e pela coluna sobreviveu e conte a quantidade de pessoas que sobreviveu de acordo com cada classe.
#Quem sobreviveu mais? A classe mais alta ou a mais baixa?
#df[['classe', 'sobreviveu']].groupby(['classe']).count()
df3 = df.copy()
def classifica_passageiro(row_value):
    if row_value == 3:
        social = 'terceira classe'
    elif row_value == 2:
        social = 'segunda classe'
    else:
        social = 'primeira classe'
    return social

df3['tipo_social'] = df['classe'].apply(classifica_passageiro)
df3[['tipo_social', 'classe', 'sobreviveu']].groupby(['tipo_social', 'sobreviveu']).count()
#Sobreviveram mais pessoas
#primeira classe	136
#segunda classe	     87
#terceira classe	119
```

```python
#Questão 12
#Agrupe pelo sexo e pela coluna sobreviveu e conte a quantidade de pessoas que sobreviveram de acordo com o sexo.
#Quem sobreviveu mais? Homens ou mulheres?
df[['id_passageiro', 'sexo', 'sobreviveu']].groupby(['sexo', 'sobreviveu']).count()
#Sobreviveram mais mulheres do que homens:
#mulheres           233
#homens             109
```
